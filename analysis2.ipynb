```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from termcolor import colored
```

```python
def read_file(file):
    data= np.loadtxt('QEEG_data/'+file+'.csv', skiprows=1, delimiter=',', unpack= True, usecols=(1,2,3,4))
    return data
```

```python
channel= np.loadtxt('QEEG_data/A1/A11.csv', skiprows=1, delimiter=',', unpack= True, usecols=(0), dtype= 'str')
```

```python
# no: 第幾次做的EEG
# cha：頻道
# fre: 頻率的index
def ch_data(chan, freq, no, mask):
    ch= np.array([])
    for i in range(1,6):
        filepath= str(mask)+str(i)+'/'+str(mask)+str(i)+str(no)
        A_ch1= read_file(filepath)[freq][chan]
        ch= np.append(ch, A_ch1)
    return ch

print('第一次')

def Wilcoxon_ch(freq, no1, no2, mask):
    p_value_w= np.array([])
    p_value_p= np.array([])
    for i in range(0, 19):
        print('Channel:', channel[i])
        A_0to4_ch1_1= ch_data(i,freq,no1, mask)
        A_0to4_ch1_2= ch_data(i,freq,no2, mask)
        print('Wilcoxon')
        a= stats.wilcoxon(A_0to4_ch1_1, A_0to4_ch1_2)
        print(a)
        print('Paired T-test')
        b= stats.ttest_rel(A_0to4_ch1_1, A_0to4_ch1_2)
        print(b)
        print('')
        p_value_w= np.append(p_value_w, a[1])
        p_value_p= np.append(p_value_p, b[1])
    return p_value_w,  p_value_p


print('面膜A 前和中比較')
print('Frequency: 0-4 Hz')
print('')
A_delta_12= Wilcoxon_ch(0, 1, 2, 'A')
print('')

        
print('面膜A 前和中比較')
print('Frequency: 4-8 Hz')
print('')
A_theta_12= Wilcoxon_ch(1, 1, 2, 'A')


print('面膜A 前和中比較')
print('Frequency: 8-13 Hz')
print('')
A_alpha_12= Wilcoxon_ch(2, 1, 2, 'A')


print('面膜A 前和中比較')
print('Frequency: 13-30 Hz')
print('')
A_beta_12= Wilcoxon_ch(3, 1, 2, 'A')  
```

```python
print('面膜A 中和後比較')
print('Frequency: 0-4 Hz')
print('')
A_delta_23= Wilcoxon_ch(0, 2, 3, 'A')
print('')

        
print('面膜A 中和後比較')
print('Frequency: 4-8 Hz')
print('')
A_theta_23= Wilcoxon_ch(1, 2, 3, 'A')


print('面膜A 中和後比較')
print('Frequency: 8-13 Hz')
print('')
A_alpha_23= Wilcoxon_ch(2, 2, 3, 'A')


print('面膜A 中和後比較')
print('Frequency: 13-30 Hz')
print('')
A_beta_23= Wilcoxon_ch(3, 2, 3, 'A')
```

```python
def P_value(List):
    for i in range(4):
        print(title[i])
        index_w= np.where(List[i][0] < 0.1)
        if len(index_w[0])!=0:
            print('Wilcoxon: p-value 小於 0.1')
            print(channel[index_w])
       
        index_p= np.where(List[i][1] < 0.1)
        if len(index_p[0])!=0:
            print('Paired T test: p-value 小於 0.1')
            print(channel[index_p])
        print('')
        

List_A12= [A_delta_12, A_theta_12, A_alpha_12, A_beta_12]
List_A23= [A_delta_23, A_theta_23, A_alpha_23, A_beta_23]
List_A13= [A_delta_13, A_theta_13, A_alpha_13, A_beta_13]

title= ['Delta', 'Theta', 'Alpha', 'Beta']

print('A面膜第一次前中比較')
P_value(List_A12)

print('A面膜第一次中後比較')
P_value(List_A23)

print('A面膜第一次前後比較')
P_value(List_A13)
```
